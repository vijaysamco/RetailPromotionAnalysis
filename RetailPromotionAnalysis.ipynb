# Retail Promotion Analysis
## Imports and Data Loading

```
import pandas as pd
import matplotlib.pyplot as plt

campaigns=pd.read_csv('datasets/dim_campaigns.csv')
products=pd.read_csv('datasets/dim_products.csv')
stores=pd.read_csv('datasets/dim_stores.csv')
events=pd.read_csv('datasets/fact_events.csv')
```

## Q1

```
df=stores['city'].value_counts().sort_values(ascending=False)
plt.figure(figsize=(8,5));df.plot(kind='bar');plt.title('Stores by City');plt.show()
```

## Q2

```
df=events.merge(campaigns,on='campaign_id').merge(products,on='product_code')
s=df[df['campaign_name'].str.contains('Sankranti',case=False)]
cat=s.groupby('category')['quantity_sold(after_promo)'].sum();plt.figure(figsize=(6,6));plt.pie(cat,labels=cat.index,autopct='%1.1f%%');plt.show()
```

## Q3

```
corr=events[['base_price','quantity_sold(after_promo)']].corr();plt.imshow(corr);plt.colorbar();plt.xticks([0,1],corr.columns,rotation=20);plt.yticks([0,1],corr.index);plt.show()
```

## Q4

```
df=events.merge(products,on='product_code')
for c in df['category'].unique():
 plt.figure();plt.hist(df[df['category']==c]['quantity_sold(before_promo)']);plt.title(c);plt.show()
```

## Q5

```
df=events.merge(stores,on='store_id')
g=df.groupby('city').agg({'quantity_sold(before_promo)':'sum','quantity_sold(after_promo)':'sum'})
g['ISU%']=(g.iloc[:,1]-g.iloc[:,0])/g.iloc[:,0]*100
plt.figure(figsize=(8,5));plt.plot(g.index,g['ISU%'],marker='o');plt.xticks(rotation=45);plt.show()
```

## Q6

```
df=events.merge(stores,on='store_id');df=df[df.city=='Hyderabad']
df['rev_before']=df['base_price(before_promo)']*df['quantity_sold(before_promo)']
df['rev_after']=df['base_price(after_promo)']*df['quantity_sold(after_promo)']
g=df.groupby('promo_type').sum(numeric_only=True)
g['ISU%']=(g['quantity_sold(after_promo)']-g['quantity_sold(before_promo)'])/g['quantity_sold(before_promo)']*100
g['IR%']=(g['rev_after']-g['rev_before'])/g['rev_before']*100
plt.scatter(g['ISU%'],g['IR%']);
[plt.text(x,y,l) for l,x,y in zip(g.index,g['ISU%'],g['IR%'])];plt.show()
```

## Q7

```
df=events.merge(stores,on='store_id').merge(products,on='product_code');df=df[df.city=='Bengaluru']
df['rev_before']=df['base_price(before_promo)']*df['quantity_sold(before_promo)']
df['rev_after']=df['base_price(after_promo)']*df['quantity_sold(after_promo)']
g=df.groupby('category')[['rev_before','rev_after']].sum();g.plot(kind='bar');plt.show()
```
